The below function aims to exploit all values in a nested dictionary, thereby summarising it into nested key/value structures and consequently giving us a list type inference.

```python
def extract_all_values(dict_temp, dimension):
    value_list_def = list(value for key, value in dict_temp.items())
    key_list_def = list(key for key, value in dict_temp.items())
    result = key_list_def if dimension == "key" else value_list_def
    return result
```

```python
# import plotly.subplots as sp

# fig = sp.make_subplots(rows=2, cols=2, subplot_titles=("Plot 1", "Plot 2", "Plot 3", "Plot 4"), specs=[[{"type": "xy"},{"type": "xy"}], [{"type": "xy"},{"type": "pie"}]])

# # Add traces to the subplots
# fig.add_trace(go.Scatter(x=[1, 2, 3], y=[4, 5, 6]), row=1, col=1)
# fig.add_trace(go.Bar(x=[1, 2, 3], y=[7, 8, 9]), row=1, col=2)
# fig.add_trace(go.Line(x=[1, 2, 3], y=[10, 11, 12]), row=2, col=1)

# # Create a new subplot of type `pie`
# pie_subplot = sp.subplots.make_subplots(rows=1, cols=1, subplot_titles=("Pie Chart"), specs=[[{"type": "pie"}]])

# # Add the pie chart trace to the new subplot
# pie_subplot.add_trace(go.Pie(labels=["A", "B", "C"], values=[10, 20, 30]), row=1, col=1)

# # Add the new subplot to the figure
# fig.add_layout_row(pie_subplot)

# # Show the plot
# fig.show()
# Use code with caution. Learn more
# I hope this helps!

# profile picture
# from datetime import datetime
# from plotly.subplots import make_subplots
# import plotly.graph_objects as go

# def plot_time_series(name_nhs_center):
#     number_of_new_positive = []
#     number_of_new_deaths = []
#     number_of_first_infections = []
#     number_of_reinfections = []
#     input_dictionary = covid_days

#     for part_dictionary in input_dictionary:
#         if part_dictionary['locations'][name_nhs_center] == {}:
#             number_of_new_positive.append(0)
#             number_of_new_deaths.append(0)
#             number_of_first_infections.append(0)
#             number_of_reinfections.append(0)
#         else:
#             number_of_new_positive.append(part_dictionary['locations'][name_nhs_center]['new_positive'])
#             number_of_new_deaths.append(part_dictionary['locations'][name_nhs_center]['new_deaths'])
#             number_of_first_infections.append(part_dictionary['locations'][name_nhs_center]['first_infections'])
#             number_of_reinfections.append(part_dictionary['locations'][name_nhs_center]['reinfections'])
    
    
    
    
#     fig = make_subplots(rows=5, cols=1, 
#                         vertical_spacing=0.1,
#                          specs = [[{"type": "xy"},{"type": "xy"}, {"type": "xy"},{"type": "xy"},{"type":"pie"}]])
#     budget_count_specific = list(chain(*list(extract_all_values(i['budget_billions'], "values") for i in boards_info if i['name'] == name_nhs_center)))
#     years = list(chain(*list(extract_all_values(i['budget_billions'], "keys") for i in boards_info if i['name'] == name_nhs_center)))
#     fig_pi = go.Figure(data = [go.Pie(labels= years, values = budget_count_specific, hole=.3)])

#     # data_sunburst = dict(
#     # budget_count = budget_count,
#     # regions = regions ,
#     # years = years
#     # )

#     # figsun = px.sunburst(data_sunburst,path= ['regions', 'years'], values = 'budget_count',color = 'budget_count', title = "SunBurst Chart of Accumulated Budgets over years for different NHS bases")

    
#     fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_new_positive, name="Number of New Positive Cases"), row=1, col=1)
#     fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_new_deaths, name="Number Of New Deaths"), row=2, col=1)
#     fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_first_infections, name="Number of First Infections Cases"), row=3, col=1)
#     fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_reinfections, name="Number of Reinfections"), row=4, col=1)
#     fig.add_trace(fig_pi.data[0],row=2,col=2)
    
#     fig.update_yaxes(title_text="Count", row=1, col=1)
#     fig.update_yaxes(title_text="Count", row=2, col=1)
#     fig.update_yaxes(title_text="Count", row=3, col=1)
#     fig.update_yaxes(title_text="Count", row=4, col=1)
    
#     fig.update_xaxes(title_text="Date", row=1, col=1)
#     fig.update_xaxes(title_text="Date", row=2, col=1)
#     fig.update_xaxes(title_text="Date", row=3, col=1)
#     fig.update_xaxes(title_text="Date", row=4, col=1)

#     fig.update_layout(height=1200, width=1000, title_text=str(name_nhs_center) + ": Time series graphs")

#     fig.show()
    
   
   
# View other drafts




import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
from plotly.subplots import make_subplots


def plot_time_series(name_nhs_center):
    """Plots a time series graph with a pie chart for a given NHS center.

    Args:
        name_nhs_center (str): The name of the NHS center.
    """

    # Get the data for the NHS center
    number_of_new_positive = []
    number_of_new_deaths = []
    number_of_first_infections = []
    number_of_reinfections = []
    input_dictionary = covid_days

    for part_dictionary in input_dictionary:
        if part_dictionary['locations'][name_nhs_center] == {}:
            number_of_new_positive.append(0)
            number_of_new_deaths.append(0)
            number_of_first_infections.append(0)
            number_of_reinfections.append(0)
        else:
            number_of_new_positive.append(part_dictionary['locations'][name_nhs_center]['new_positive'])
            number_of_new_deaths.append(part_dictionary['locations'][name_nhs_center]['new_deaths'])
            number_of_first_infections.append(part_dictionary['locations'][name_nhs_center]['first_infections'])
            number_of_reinfections.append(part_dictionary['locations'][name_nhs_center]['reinfections'])

    # Create a subplots figure
    fig = make_subplots(rows=2, cols=2, vertical_spacing=0.1, specs=[[{"type": "xy"}, {"type": "xy"}], [{"type": "xy"}, {"type": "pie"}]])

    # Add a trace for each date
    for i in range(len(dates_covid_data)):
        fig.add_trace(go.Scatter(x=[dates_covid_data[i]], y=[number_of_new_positive[i]], name="Number of New Positive Cases"), row=1, col=1)
        fig.add_trace(go.Scatter(x=[dates_covid_data[i]], y=[number_of_new_deaths[i]], name="Number Of New Deaths"), row=1, col=2)
        #fig.add_trace(go.Scatter(x=[dates_covid_data[i]], y=[number_of_first_infections[i]], name="Number of First Infections Cases"), row=1, col=1)
        fig.add_trace(go.Scatter(x=[dates_covid_data[i]], y=[number_of_reinfections[i]], name="Number of Reinfections"), row=2, col=1)

    # Add a pie chart for the budget
    budget_count_specific = list(chain(*list(extract_all_values(i['budget_billions'], "values") for i in boards_info if i['name'] == name_nhs_center)))
    years = list(chain(*list(extract_all_values(i['budget_billions'], "keys") for i in boards_info if i['name'] == name_nhs_center)))

    fig_pi = go.Figure(data=[go.Pie(labels=years, values=budget_count_specific, hole=.3)])
    fig.add_trace(fig_pi.data[0], row=2, col=2)

    # Update the layout of the figure
    fig.update_layout(height=1200, width=1000, title_text=str(name_nhs_center) + ": Time series graphs with a pie chart")

    # Show the figure
    fig.show()
```

```python
from datetime import datetime
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def plot_time_series(name_nhs_center):
    number_of_new_positive = []
    number_of_new_deaths = []
    number_of_first_infections = []
    number_of_reinfections = []
    input_dictionary = covid_days

    for part_dictionary in input_dictionary:
        if part_dictionary['locations'][name_nhs_center] == {}:
            number_of_new_positive.append(0)
            number_of_new_deaths.append(0)
            number_of_first_infections.append(0)
            number_of_reinfections.append(0)
        else:
            number_of_new_positive.append(part_dictionary['locations'][name_nhs_center]['new_positive'])
            number_of_new_deaths.append(part_dictionary['locations'][name_nhs_center]['new_deaths'])
            number_of_first_infections.append(part_dictionary['locations'][name_nhs_center]['first_infections'])
            number_of_reinfections.append(part_dictionary['locations'][name_nhs_center]['reinfections'])
    
           
    fig = make_subplots(rows=2, cols=2, vertical_spacing=0.1, specs=[[{"type": "xy"}, {"type": "xy"}], [{"type": "xy"}, {"type": "sunburst"}]])
    budget_count_specific = list(chain(*list(extract_all_values(i['budget_billions'], "values") for i in boards_info if i['name'] == name_nhs_center)))
    years_specific = list(chain(*list(extract_all_values(i['budget_billions'], "keys") for i in boards_info if i['name'] == name_nhs_center)))
    fig_pi = go.Figure(data = [go.Pie(labels= years, values = budget_count_specific, hole=.3, textinfo = 'label')])
    data_sunburst = dict(
    budget_count = budget_count,
    regions = list(i for i in regions for _ in range(3)) ,
    years = years,
    staff_per_area_km2 = list(i for i in staff_per_area for _ in range(3))
    )


    figsun = px.sunburst(data_sunburst,
                         path= ['regions', 'years'], 
                         values = 'budget_count',
                         hover_data= ['staff_per_area_km2'],
                         title = "SunBurst Chart of Accumulated Budgets over years for different NHS bases",
                        )

    # figsun.add_trace(hovertemplate='<b>%{label} </b> <br> Sales: %{value}<br> Success rate: %{color:.2f}',)
    fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_new_positive, name="Number of New Positive Cases"), row=1, col=1)
    fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_new_deaths, name="Number Of New Deaths"), row=2, col=1)
    fig.add_trace(go.Scatter(x=dates_covid_data, y=number_of_first_infections, name="Number of First Infections Cases"), row=1, col=2)
    fig.add_trace(figsun.data[0],row=2,col=2)
    fig.update_traces(customdata = list(staff_per_area*4) ,hovertemplate='<b>%{label} </b> <br> Staff per square km: %{customdata}<br> Budget in billions : %{value}', row = 2, col= 2)
    fig.update_yaxes(title_text="Count of new positive Cases", row=1, col=1)
    fig.update_yaxes(title_text="Count of Deaths due to Covid", row=2, col=1)
    fig.update_yaxes(title_text="Count of Covid-19 reinfections", row=1, col=2)
    fig.update_coloraxes(title_text="Budget Split over years and NHS Zones", row=2, col=2)
    
    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_xaxes(title_text="Date", row=1, col=2)

    fig.update_layout(height=1200, width=1500, title_text= str(name_nhs_center) + ": Time series graphs and Sunburst chart of Budget distribution over years and NHS zones")

    fig.show()
```
